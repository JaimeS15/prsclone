```python
import pandas as pd
```

# Read in files
Note: trace data only has information from 2019

```python
loc1 = pd.read_csv("trace_data/location_pt1.csv")
loc2 = pd.read_csv("trace_data/location_pt2.csv")
loc3 = pd.read_csv("trace_data/location_pt3.csv")
loc4 = pd.read_csv("trace_data/location_pt4.csv")
loc5 = pd.read_csv("trace_data/location_pt5.csv")
loc6 = pd.read_csv("trace_data/location_pt6.csv")
```

```python
trips = pd.read_csv("Household_Travel_Survey_Trips.csv")
persons = pd.read_csv("Household_Travel_Survey_Persons.csv")
```

```python
# combine all lat/lon info
all_loc = pd.concat([loc1, loc2, loc3, loc4, loc5, loc6])
```

# Enrich lat/lon dataset with person and trip info
Combine trip info through personid and trip arrival/departure time?

```python
loc_info = pd.merge(all_loc, persons, how="left", left_on="personid", right_on="person_id")
```
